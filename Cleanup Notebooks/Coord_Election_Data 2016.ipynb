```python
#Dependencies
import pandas as pd
import numpy as np
import os
```

```python
elec16_file = "Data/elec16v2.csv"
elec20_file = "Data/elec20v2.csv"
coord_file = "Data/us_county_latlng.csv"

coord_data = pd.read_csv(coord_file, low_memory=False)
elec20_data = pd.read_csv(elec20_file, low_memory=False)
elec16_data = pd.read_csv(elec16_file, low_memory=False)
```

```python
coord_data.dtypes
```

```python
elec16_data.dtypes
```

```python
coord_data = coord_data.rename(columns = {"fips_code":"countyfips"})
```

```python
coord_data
```

```python
map_data = pd.merge(elec16_data, coord_data, how="left", on=["countyfips"])
map_data
```

```python
map_data_final = map_data.query("countyfips != 0")
map_data_final
```

```python
map_data_grouped = map_data_final.groupby(['combinedfips','countyfips','statefips','party_simple',"lng","lat"])['votes'].sum()
```

```python
map_data_grouped
```

```python
map_data_grouped.to_csv('Data/map_data_noprimary.csv')
```

```python
map_data_final2 = map_data2.query("countyfips != 0")
map_data_final2
```
